```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

results1 = np.loadtxt(open("experiment_1/exp1_saveddata.csv", "rb"), delimiter=",", skiprows=1)
results2 = np.loadtxt(open("experiment_1/exp1_grid_UCB.csv", "rb"), delimiter=",", skiprows=1)
results3 = np.loadtxt(open("exp1_EG.csv", "rb"), delimiter=",", skiprows=1)
results4 = np.loadtxt(open("exp2_with_EG_too_with_25_reps/exp2_grid_UCB.csv", "rb"), delimiter=",", skiprows=1)
results5 = np.loadtxt(open("exp2_with_EG_too_with_25_reps/exp2_leftist.csv", "rb"), delimiter=",", skiprows=1)
results6 = np.loadtxt(open("exp2_EG.csv", "rb"), delimiter=",", skiprows=1)
results7 = np.loadtxt(open("experiment_3_tau_point2_p1_point55_p0_point25/exp3_grid_UCB.csv", "rb"), delimiter=",", skiprows=1)
results8 = np.loadtxt(open("experiment_3_tau_point2_p1_point55_p0_point25/exp3_saveddata.csv", "rb"), delimiter=",", skiprows=1)
results9 = np.loadtxt(open("exp3_EG.csv", "rb"), delimiter=",", skiprows=1)
results10 = np.loadtxt(open("tau_point1_p1_point3_p0_point25/exp3_grid_UCB.csv", "rb"), delimiter=",", skiprows=1)
results11 = np.loadtxt(open("tau_point1_p1_point3_p0_point25/exp3_saveddata.csv", "rb"), delimiter=",", skiprows=1)
results12 = np.loadtxt(open("exp4_EG.csv", "rb"), delimiter=",", skiprows=1)
# Make sure x and reps match the ones that are set in TLB_Exp.py 
# when you collected the experimental results

reps=25
x=40
T=1000000
x2=30

# Leftist Exp 1
df1 = pd.DataFrame(results1[:,5])
leftist_exp1 = ((df1.groupby(np.arange(len(df1))//reps).mean())/10000).to_numpy()

# Grid UCB Exp 1
df2 = pd.DataFrame(results2[:,2])
GUCB_exp1 = ((df2.groupby(np.arange(len(df2))//reps).mean())/10000).to_numpy()

# EG Exp 1
df = pd.DataFrame(results3[:,5])
EG_exp1 = ((df.groupby(np.arange(len(df))//reps).mean())/10000).to_numpy()

taulin = np.linspace(0.01,0.4,x)

static_delta = 0.1
static_tau = 0.1


plt.rcParams['xtick.labelsize']=12
plt.rcParams['ytick.labelsize']=12

# Experiment 1 plot    
plt.figure(1)
plt.plot(taulin,leftist_exp1,color= 'blue', label = "Leftist", linewidth = 2)
plt.plot(taulin,GUCB_exp1,color= 'orange', label = "Grid UCB", linewidth = 2)
plt.plot(taulin, EG_exp1, color= 'black', label = "EG", linewidth = 2)
plt.vlines(static_delta,0,np.max(EG_exp1),colors="red",linewidth=1, label = "$\Delta$")
plt.xlabel(r"$\tau$", fontsize=18)
plt.ylabel("Pseudo-regret $\cdot 10^4$", fontsize= 18)
plt.legend(loc='best', bbox_to_anchor=(1, 0.75), fontsize='large')
plt.savefig("Exp1_plot_EG.pdf", format="pdf")
```

```python
# EG Exp 2
eg = pd.DataFrame(results6[:,5])
EG_exp2 = ((eg.groupby(np.arange(len(eg))//reps).mean())/1000000).to_numpy()

# Leftist Exp 2
df3 = pd.DataFrame(results5[:,5])
leftist_exp2 = ((df3.groupby(np.arange(len(df3))//reps).mean())/1000000).to_numpy()

# Grid UCB Exp 2
df4 = pd.DataFrame(results4[:,2])
GUCB_exp2 = ((df4.groupby(np.arange(len(df4))//reps).mean())/1000000).to_numpy()
delta = np.geomspace(5e-3, 0.55, x2)    
```

```python
# Experiment 2 plot
plt.figure(2)
plt.plot(delta,leftist_exp2,color="blue", label = "Leftist", linewidth=2)
plt.plot(delta,GUCB_exp2,color="orange", label = "Grid UCB", linewidth = 2)
plt.plot(delta, EG_exp2, color= 'black', label = "EG", linewidth = 2)
plt.vlines(static_tau,0,np.max(EG_exp2),colors="red",linewidth=1, label = r"$\tau$")
plt.xlabel("$\Delta$", fontsize=18)
plt.xscale("log")
plt.ylabel("Pseudo-regret $\cdot 10^6$",fontsize=18)
plt.legend(loc='best', fontsize='large')
plt.savefig("Exp2_plot_EG.pdf", format="pdf")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# UCB Exp 3
ucb7 = pd.DataFrame(results7[:,2])
GUCB_exp3 = ((ucb7.groupby(np.arange(len(ucb7))//reps).mean())/100000).to_numpy()

# Leftist Exp 3
l3 = pd.DataFrame(results8[:,5])
leftist_exp3 = ((l3.groupby(np.arange(len(l3))//reps).mean())/100000).to_numpy()

# EG Exp 3
eg3 = pd.DataFrame(results9[:,5])
EG_exp3 = ((eg3.groupby(np.arange(len(eg3))//reps).mean())/100000).to_numpy()

# UCB Exp 3
ucb10 = pd.DataFrame(results10[:,2])
GUCB_exp32 = ((ucb10.groupby(np.arange(len(ucb10))//reps).mean())/100000).to_numpy()

# Leftist Exp 3
lf3 = pd.DataFrame(results11[:,5])
leftist_exp32 = ((lf3.groupby(np.arange(len(lf3))//reps).mean())/100000).to_numpy()

# EG Exp 3
eg32 = pd.DataFrame(results12[:,5])
EG_exp32 = ((eg32.groupby(np.arange(len(eg32))//reps).mean())/100000).to_numpy()

reps=25
T=1000000
x3=6

time = np.geomspace(5e5, 16e6, x3, dtype=int)
```

```python
plt.figure(4)
plt.plot(time,leftist_exp3,color= 'blue', label = r"Leftist ($\Delta > \tau $)")
plt.plot(time,leftist_exp32,color= 'blue', linestyle='dashed', label = r"Leftist ($\tau > \Delta$)")
plt.plot(time,GUCB_exp3,color= 'orange', label = r"Grid UCB ($\Delta > \tau $)")
plt.plot(time,GUCB_exp32,color= 'orange', linestyle='dashed', label = r"Grid UCB ($\tau > \Delta$)")
plt.plot(time,EG_exp3,color= 'black', label = r"EG ($\Delta > \tau $)")
plt.plot(time,EG_exp32,color= 'black', linestyle='dashed', label = r"EG ($\tau > \Delta$)")
plt.xlabel("T", fontsize=18)
plt.ylabel("Pseudo-regret $\cdot 10^5$", fontsize=18)
plt.legend(loc="best", bbox_to_anchor=(1, 0.7))
plt.savefig("time_plot_EG.pdf", bbox_inches="tight", format="pdf")
```
